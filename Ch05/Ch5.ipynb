### Some basic inputs

```python
from numpy import *
%matplotlib inline
import matplotlib.pyplot as plt
```

### Sigmoid Function

```python
def sigmoid(x):
    return 1.0/(1.0 + exp(-x))

x=array(arange(-10, 10, 0.1))
y=sigmoid(x)
plt.figure()
plt.plot(x,y)
plt.show()
```

### Sigmoid of f(x)

```python
x=array(arange(-2, 2, 0.01))
y=3*x**3
y=sigmoid(y)
plt.figure()
plt.plot(x,y)
plt.show()
```

### Linear Cost Function

```python
theta=array(arange(-5, 5, 0.1))

import random as rand
x=array([rand.choice(arange(-6, 6, 0.01)) for i in arange(100)])
y=array([1 if item ** 3 + item > 20 else 0 for item in x])

print(x)
print(y)

def linearCost(theta, x, y):
    cost = zeros(len(theta))
    for i in arange(len(x)):
        sig = sigmoid(theta * x[i])
        cost += (sig - y[i]) ** 2 / 2.0
    return cost

def logisticCost(theta, x, y):
    cost = zeros(len(theta))
    for i in arange(len(x)):
        sig = sigmoid(theta * x[i])
        cost -= y[i] * log(sig) + (1 - y[i]) * log(1 - sig)
    return cost

plt.figure()
plt.plot(theta,linearCost(theta, x ,y))
plt.show()
plt.figure()
plt.plot(theta,logisticCost(theta, x ,y))
plt.show()
```

### Data set

```python
def loadDataSet():
    dataSet=[]
    labelSet=[]
    with open('testSet.txt', 'r') as tf:
        for line in tf:
            oneData=line.strip().split()
            dataSet.append([float(oneData[0]), float(oneData[1])])
            labelSet.append(int(oneData[2]))
    return dataSet,labelSet

def showDataSet(dataSet, labelSet):
    #print some data for review
    for i in range(10):
        print(dataSet[i][0], dataSet[i][1], labelSet[i])
    
    # picture
    type0_x=[]
    type0_y=[]
    type1_x=[]
    type1_y=[]
    for i in range(len(labelSet)):
        if labelSet[i]==0:
            type0_x.append(dataSet[i][0])
            type0_y.append(dataSet[i][1])
        elif labelSet[i]==1:
            type1_x.append(dataSet[i][0])
            type1_y.append(dataSet[i][1])

    plt.figure()
    plt.scatter(type0_x, type0_y, c='red', s=20)
    plt.scatter(type1_x, type1_y, c='blue', s=20)
    plt.show()
    
dataSet,labelSet = loadDataSet()
showDataSet(dataSet, labelSet)
```
